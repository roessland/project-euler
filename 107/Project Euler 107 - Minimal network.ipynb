## Parse input

0 represents infinite cost (no edge) in a sparse matrix.

```python
graph = csr_matrix([[0 if num=="-" else int(num) for num in line.split(",")] for line in network_txt.split("\n")])
print(graph.todense())
```

## Create minimum spanning tree

```python
mst = minimum_spanning_tree(graph)
mst_directed = mst.todense().T + mst.todense()
```

## Find total savings

```python
# Edges are counted both ways, so divide by two
np.sum(graph)/2 - np.sum(mst_directed)/2
```
